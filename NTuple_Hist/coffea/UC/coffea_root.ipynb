```python
import ROOT as r
```

```python
f_coffea = r.TFile("coffea.root", "RO")
h_coffea = f_coffea.Get("all")
f_frames = r.TFile("output/example_FS.root", "RO")
h_frames = f_frames.Get("NOSYS/ph_pt_Muon")
c_weighted = r.TCanvas("c_weighted", "c_weighted", 800, 600)
```

```python
c_weighted.Draw()
c_weighted.SetLogy()
h_frames.SetLineColor(r.kRed)

h_frames.Draw()
h_frames.GetYaxis().SetRangeUser(100, 1e6)
h_coffea.Draw("sames")
```

```python
print(h_coffea.GetEntries())
print(h_frames.GetEntries())

print(h_coffea.GetSumOfWeights())
print(h_frames.GetSumOfWeights())
```

```python
c_unweighted = r.TCanvas("c_unweighted", "c_unweighted", 800, 600)
c_unweighted.SetLogy(1)

c_unweighted.Draw()
h_frames.Draw()
h_coffea.Draw("same")
```

```python
c_check = r.TCanvas("c_check", "c_check", 800, 600)

c_check.Draw()
h_check = h_frames.Clone()
h_check.Divide(h_coffea)
h_check.GetYaxis().SetRangeUser(0, 2)
h_check.Draw()
```
