```python
import ROOT as r
```

```python
f_coffea=r.TFile("/data/selbor/ntuple/coffea/coffea.root","RO")
h_coffea=f_coffea.Get("all")
f_frames=r.TFile("/data/selbor/ntuple/fastframes/output/example_FS_single.root","RO")
h_frames=f_frames.Get("NOSYS/ph_pt_Muon")
c_weighted=r.TCanvas("c_weighted","c_weighted",800,600)
```

```python
c_weighted.Draw()
c_weighted.SetLogy()
h_frames.SetLineColor(r.kRed)

h_frames.Draw()
h_frames.GetYaxis().SetRangeUser(100,1e22)
h_coffea.Draw("same")
```

```python
print(h_coffea.GetEntries())
print(h_frames.GetEntries())

print(h_coffea.GetSumOfWeights())
print(h_frames.GetSumOfWeights())
```
